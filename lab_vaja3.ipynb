```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
```

```python
x = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 52])*0.001 #seznam vrednosti radija cevi v mm
v1 = np.array([1.49, 1.76, 1.69, 1.68, 1.49, 1.26, 1.05, 0.99, 0.86, 0.71, 0.65]) #seznam izmerjenih vrednosti hitrosti toka v m/s, od r = -52mm do r = -5mm
v1 = v1[::-1] #obratno sortiranje seznama v1
v2 = [0.55] #hitrost v središču cevi (r = 0mm)
v3 = np.array([0.75, 0.84, 0.98, 1.22, 1.24, 1.46, 1.66, 1.65, 1.52, 1.34, 1.34]) #seznam izmerjenih vrednosti hitrosti toka v m/s, od r = 5mm do r = 52mm
for i, j in zip(v1, v3): #izračun povprečne hitrosti toka v cevi
    v2.append((i+j)/2)
v = np.array(v2) # pretvorba seznama povprečnih vrednosti hitrosti v numpy array

plt.plot(x, v, "o-", c="r") #izris grafa
plt.xlabel("radij cevi r [m]")
plt.ylabel("hitrost toka v [m/s]")
plt.show()
```

```python
i = np.polyfit(x, v, 3) #aproksimacija profila hitrosti po radiju s polinomom 3. stopnje
p = np.poly1d(i) #izračun polinoma
x2 = np.linspace(0, 0.052, 101) 
plt.plot(x2, p(x2), c = "r", label = "aproksimacija profila hitrosti po radiju s polinomom 3. stopnje") #izris grafa aproksimacije
plt.xlabel("radij cevi r [m]")
plt.ylabel("hitrost toka v [m/s]")
plt.legend()
plt.show()
```

```python
test = abs(x2 * p(x2)) # izračun vrednosti v(r) * r
int = 2 * np.pi * spi.simpson(y = test, x = x2) * 1e3 # integracija vrednosti v(r) * r po radiju in pretvorba v l/s
print(f"Pretok skozi cev je {int:.2f} l/s")
```
